```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
X = df.drop("prognosis",axis=1)  #dropping it column-wise, if row wise then axis=0 #represents input variablee(independent)
y = df["prognosis"] #represents output varible (dependent)
```

```python
le = LabelEncoder()
le.fit(y)
Y = le.transform(y)
#encoding categorical values into integer values
```

```python
Y
```

```python
X
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,Y,test_size= 0.3, random_state= 20)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape 
#the model is divided into trainig=70% , testing=30%
```

```python
from sklearn.datasets import make_classification
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np

#storing the models in a dictionary and checking which is most suiltable
models = {
    "SVC":SVC(kernel='linear'),
    "RandomForest" : RandomForestClassifier(n_estimators= 100, random_state= 42),
    "KNeighbors": KNeighborsClassifier(n_neighbors= 5),

}

for model_name, model in models.items():
    #training model
    model.fit(X_train, y_train)
    #testing
    predictions = model.predict(X_test)
    #calculation accuracy
    accuracy = accuracy_score(y_test, predictions)
    #confusion matrix 
    cm = confusion_matrix(y_test, predictions)

    print(f"{model_name} accuracy : {accuracy}")
    print(f"{model_name} Confusion matrix: ")
    print(np.array2string(cm, separator= ','))
    
    
```

```python
#selecting the svm model

svc = SVC(kernel='linear')
#training
svc.fit(X_train, y_train)
#testing
ypred= svc.predict(X_test)
#calculation accuracy
accuracy_score(y_test, ypred)
```
